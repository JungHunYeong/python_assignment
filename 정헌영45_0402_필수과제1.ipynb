##1. append

```python
# append: 시리즈에 다른 시리즈나 데이터프레임을 추가
newdata = pd.Series(['new'], index = ['NEW INDEX']) 
mpg.cty.append(newdata) #mpg.cty Series에 newdata라는 시리즈를 추가해 붙인다
```

## 2. describe

```python
# describe: 시리즈의 기초 통계량을 요약해서 보여줌
mpg.cty.describe()
```

## 3. drop_duplicates

```python
## drop_duplicates: 시리즈에서 중복된 값을 제거
mpg.cty.drop_duplicates() #unique한 값만 나타나는것을 확인할 수 있음
```

## 4. get_values

```python
#get_values: 시리즈의 값들을 넘파이 배열로 반환
mpg.cty.get_values() #오류: get_values() 메서드는 더이상 사용되지 않음. 대신 to_numpy 사용
```

```python
mpg.cty.to_numpy
```

##5. isin

```python
#isin: 시리즈의 값들이 주어진 값인지 여부를 반환
mpg.cty.isin([np.nan]) #각 값들이 NaN인지 여부를 반환하는 시리즈 출력됨
```

## 6. 기초통계량(mean, max, min, sum, var)

```python
print('Mean is',mpg.cty.mean()) #mean: 평균
print('Max is', mpg.cty.max()) #max: 최댓값
print('Min is', mpg.cty.min()) #min: 최솟값
print('Sum is', mpg.cty.sum()) #sum; 합
print('Variance is',mpg.cty.var()) #var: 분산
```

## 7.equals

```python
#equals: 시리즈와 다른 시리즈나 데이터프레임이 같은지 여부를 반환
cty2 = mpg.cty.copy()
mpg.cty.equals(cty2)
```

##8. sort_values

```python
 #sort_values: 시리즈의 값을 정렬
 mpg.cty.sort_values()
```

##9. to_frame

```python
#to_frame: 시리즈를 데이터프레임으로 변환
mpg.cty.to_frame()
```
