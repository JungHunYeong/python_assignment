### 필수과제1

```python
import pandas as pd
import datetime
```

- 시계열데이터 문법 5가지

- Dateoffset

```python
# 현재 날짜를 기준으로 특정일 이후의 날짜 계산
current_date = pd.Timestamp('2023-05-26')
future_date = current_date + pd.DateOffset(weeks=1)
print(future_date)
```

- offsets.MonthBegin

```python
# 현재 날짜를 기준으로 다음달의 첫번째 날짜 출력
current_date = pd.Timestamp('2023-05-26')
offset = pd.offsets.MonthBegin()
firstday = current_date + offset
print(firstday)
```

- resample('W-MON')

```python
# W-MON: 달의 첫번째 월요일 출력
crime= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/crime.csv')

crime['REPORTED_DATE']=pd.to_datetime(crime['REPORTED_DATE'])
crime = crime.set_index('REPORTED_DATE')

first_monday = crime.resample('W-MON').first()
first_monday
```

### resample의 인자 코드:

- B: 평일(비즈니스 데이, Business day)


- D: 일일


- W: 주간(일요일을 기준으로 한 주)


- M: 월말


- Q: 분기말


- A: 연말(12월 말)



- resample('B')

```python
# 영업일만 출력
B = crime.resample('B').first()
B
```

- first

```python
#위에서 구한 영업일 데이터중 가장 첫번째 날의 데이터를 출력
B.first('D')
```
