```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
mpg = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/mpg.csv')
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

ax[0, 0].scatter(x='displ', y='cty', data=mpg)
ax[0, 1].hist(mpg['cty'])
ax[1, 0].hist(mpg['displ'])
sns.boxplot(x='cty', y='displ', data=mpg, ax=ax[1, 1])

#제목 달기
ax[0, 0].set_title('displ - cty scatter plot')
ax[0, 1].set_title('cty')
ax[1, 0].set_title('displ')
ax[1, 1].set_title('cty-displ boxplot')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
fig, ax = plt.subplots(nrows=3, ncols=3, figsize=(15, 15))

ax[0, 0].hist(titanic['fare'])
ax[0, 1].scatter(x='fare', y='age', data=titanic)
sns.boxplot(x='pclass', y='fare', data=titanic, ax=ax[0, 2])
ax[1, 1].hist(titanic['age'])
ax[1, 0].scatter(x='age', y='fare', data=titanic)
sns.boxplot(x='pclass', y='age', data=titanic, ax=ax[1, 2])
ax[2, 2].hist(titanic['pclass'])
sns.boxplot(x='sex', y='fare', data=titanic, ax=ax[2, 0])
sns.boxplot(x='sex', y='age', data=titanic, ax=ax[2, 1])

ax[0, 0].set_title('Fare')
ax[0, 1].set_title('Fare vs Age')
ax[1, 1].set_title('Age')
ax[1, 0].set_title('Age vs Fare')
ax[2, 2].set_title('Class')
```
