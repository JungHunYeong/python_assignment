### displ과 cty/hwy 관계 scatter plot

```python
#그래프 크기
plt.figure(figsize=(10,5))
sns.set(style='whitegrid')
plt.scatter(mpg.displ,mpg.cty, label = 'cty', color='blue', alpha=0.5)
plt.scatter(mpg.displ,mpg.hwy, label = 'hwy', color='green', alpha=0.5)
#제목 지정
plt.title('mpg.displ~cty/hwy')
#라벨 지정
plt.xlabel('mpg.displ') 
plt.ylabel('mpg.cty / mpg.hwy') 
#범위 지정
plt.xlim(0,8)
plt.ylim(0,50)
#눈금값 지정
plt.xticks([1,4,7],['Low(=1)','Normal(=4)','High(=7)'])
plt.legend(loc='upper right')
```
